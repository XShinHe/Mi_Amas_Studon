* Do-Block Syntax for Function Arguments 
```Julia
map(x->begin

           return 0
       elseif x == 0
           return 1
       else
           return x
       end
   end,
[A, B, C])
```  
or
```Julia
map([A, B, C]) do x
    if x < 0 && iseven(x)
        return 0
    elseif x == 0
        return 1
    else
        return x
    end
end
```
and open as example:  
```

    write(io, data)
end
```
* and `Function` can as argument: `function open(f::Function, args...)`

* verctorize function `f.(args...)`  
> * For example, `sin.(cos.(X))` is equivalent to `broadcast(x -> sin(cos(x)), X)`, similar to `[sin(cos(x)) for x in X]`  
> * `X .= ...`, which is equivalent to `broadcast!(identity, X, ...)`
> * `X[2:end] .= sin.(Y)`, then it translates to broadcast on a view, e.g. `broadcast!(sin, view(X, 2:lastindex(X)), Y)`
> * `X .+= Y` etcetera is equivalent to `X .= X .+ Y`
> * macro `@.` is provided to convert every function call, operation, and assignment in an expression into the "dotted" version. `@. X = sin(cos(Y))`
> * with function chaining using `|>` as:  
`[1:5;] .|> [x->x^2, inv, x->2*x, -, isodd]`
